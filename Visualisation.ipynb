```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

**Scatter plot**

```python
import matplotlib.pyplot as plt

plt.scatter(tips['total_bill'], tips['tip'])
# Giving our plot a title
plt.title("This is Scatter Plot")

plt.xlabel('Total bill')

plt.ylabel("Tip")
```

```python
plt.scatter(tips['day'], tips['tip'])
# Giving our plot a title
plt.title("This is Scatter Plot")

plt.xlabel('Day')

plt.ylabel("Tip")
```

```python
import plotly.express
graph = plotly.express.scatter(tips, x="total_bill", y = "tip", color = "smoker")
graph.show()
```
